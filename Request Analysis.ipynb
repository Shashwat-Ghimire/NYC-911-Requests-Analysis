**Arrangement and Graph**

```python
#Group by Borough and Complaint Type, compute mean Request_Closing_Time
avg_closing_time = data_set.groupby(['Location Type', 'Complaint Type'])['Request_Closing_Time'].mean().reset_index()
#Group by Request_Closing_Time within each Borough
avg_closing_time = avg_closing_time.sort_values(by=['Location Type', 'Request_Closing_Time'], ascending=[True, False])
pivot_data = avg_closing_time.pivot(index='Location Type', columns='Complaint Type', values='Request_Closing_Time').fillna(0)

# Create heatmap
plt.figure(figsize=(18, 8))
sns.heatmap(pivot_data, annot=True, fmt='.2f', cmap='Reds', cbar_kws={'label': 'Average Request Closing Time (Days)'}, linewidths=0.5)
plt.title('Heatmap of Average Request Closing Time by Complaint Type and Location Type', fontsize=14)
plt.xlabel('Complaint Type', fontsize=12)
plt.ylabel('Location Type', fontsize=12)
plt.xticks(rotation=35, ha='right', fontsize=10)
plt.yticks(fontsize=10)
plt.tight_layout()
plt.show()
```



**Test 1:
Whether the average response time across complaint types is similar or not**

```python
from scipy.stats import f_oneway

#grouping request closing time according to complaint type
groups = [group['Request_Closing_Time'].values for name, group in data_set.groupby('Complaint Type')]

#one-way ANOVA test
f_stat, p_value = f_oneway(*groups)

#overall mean
overall_mean = data_set['Request_Closing_Time'].mean()

#Interpretation
print("F-statistic: " + str(f_stat))
print("P-value: " + str(p_value))
print("Overall Mean Request_Closing_Time: " + str(overall_mean) + "days")
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: Mean is significantly different.")
else:
    print("Fail to reject the null hypothesis: No significant difference.")
```

**Test 2: Whether the type of complaint or service requested and location are related.**

```python
from scipy.stats import chi2_contingency

#contingency table
contingency_table = pd.crosstab(data_set['Complaint Type'], data_set['Location Type'])

#Chi-Square Test
chi2_stat, p_value, dof, expected = chi2_contingency(contingency_table)

print("\nTest 2: Chi-Square Test Results")
print("Chi-Square Statistic: " + str(chi2_stat))
print("P-value: " +str(p_value))
print("Degrees of Freedom: "+str(dof))

#Interpretation
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: Complaint Type and Location are related.")
else:
    print("Fail to reject the null hypothesis: Complaint Type and Location are independent.")
```
